# Palette Example

Generating a square wordcloud from the US constitution and using a custom palette to color it.

```python
from os import path
from wordcloud import WordCloud,  ImagePaletteGenerator
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Read the whole text.
text = open('constitution.txt').read()

# Generate a word cloud image
wordcloud = WordCloud().generate(text)
```

## Create palette from image

```python
def plot_wordcloud(wordcloud, title=''):
    fig = plt.figure(frameon=False, figsize=(19.20,10.8), dpi=100)
    ax = fig.add_axes([0, 0, 1, 1])                
    ax.set_title(title)
    ax.axis("off")
    plt.imshow(wordcloud)
    plt.show()
```

```python
# Create palette from image
palette_imagefile =  'cheeremo.jpg'
ipc_from_filename = ImagePaletteGenerator().add_image(palette_imagefile, 5)

wordcloud.recolor(color_func=ipc_from_filename)
plot_wordcloud(wordcloud, "Palette A")
```

```python
# Create palette from array
palette_image =  np.array(Image.open('florahues.jpg'))
ipc_from_array = ImagePaletteGenerator().add_image(palette_image)
wordcloud.recolor(color_func=ipc_from_array)
# Display the generated image:
# the matplotlib way:
plot_wordcloud(wordcloud,"Palette B")
```

```python
# Create palette from multiple images
palette_imagefile_emo =  'cheeremo.jpg'
palette_imagefile_flora =  'florahues.jpg'
ipc_from_multiple_files = ImagePaletteGenerator()\
    .add_image(palette_imagefile_,5)\
    .add_image(palette_imagefile_emo,5)\
    .shuffle_colors()
wordcloud.recolor(color_func=ipc_from_multiple_files)
plot_wordcloud(wordcloud, "Palette A+B Shuffled")
```
